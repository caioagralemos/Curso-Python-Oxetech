```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
df = pd.read_excel('./Relogio.xlsx')
```

```python
display(df)
```

```python
# Substituindo valores vazios
df['Duração'].fillna(df['Duração'].mean(), inplace=True)
df['Pulso médio'].fillna(df['Pulso médio'].mean(), inplace=True)
df['Pulso Minimo'].fillna(df['Pulso Minimo'].mean(), inplace=True)
df['Calorias gastas'].fillna(df['Calorias gastas'].mean(), inplace=True)
```

```python
# atualizando valores especificos
df.at[7, 'Duração'] = 45.0
df.at[22, 'Duração'] = 60.0
```

```python
df.count()
```

```python
plt.hist(df['Duração'], color='blue', alpha = 0.5)
plt.xlabel('Duração')
plt.ylabel('Frequência')
plt.title('Histograma da Duração')
plt.show()
```

```python
plt.scatter(df['Calorias gastas'], df['Pulso Minimo'], color='red', alpha = 0.5)
plt.xlabel('Calorias gastas')
plt.ylabel('Pulso Mínimo')
plt.title('Dispersão das Calorias Gastas por Pulso Mínimo')
plt.show()
```

```python
sb.boxplot(x=df['Duração'], data=df, color='red')
plt.show()
```
