```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
print(x)
print(y)

fig, ax = plt.subplots() # fig é a figura e ax são os eixos da figura
# ax.scatter(x,y) # mostra os pontos x,y na figura
# ax.plot(x,y) # plota um grafico com os pontos x,y
ax.bar(x,y, width=0.1,edgecolor="red",linewidth=0.5)
plt.show()
```

```python
ax.stem(x,y)
plt.show()
```
